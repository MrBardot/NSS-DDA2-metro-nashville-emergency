```python
def get_zip(df, lat_col, lng_col,df_zip):
    df_latlng = df[[lat_col,lng_col]].copy()
    df_latlng['Coordinates'] = list(zip(df_latlon[lon_col], df_latlon[lat_col]))
    df_latlng['Coordinates'] = df_latlon['Coordinates'].apply(Point)
    df_latlng = gpd.GeoDataFrame(df_latlng, geometry='Coordinates')
    df_latlng = df_latlon.set_crs(epsg=4326)
    df_latlng = gpd.sjoin(df_latlng, df_zip[['zip','geometry']], how='left')
    return df_latlon
```
